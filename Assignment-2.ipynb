10. Create a 2x2 matrix A and a 2x2 matrix B. Perform matrix multiplication using np.dot() or
the @ operator.

```python
import numpy as np

A = np.random.randint(1,5,4).reshape(2,2)
B=np.random.randint(1,10,4).reshape(2,2)
res1 = np.dot(A,B)
res2=A@B
print("matrix multiplication is np.dot : ",res1)
print("matrix multiplication is @ : ",res2)
```

11. Find the transpose of matrix A and matrix B.
