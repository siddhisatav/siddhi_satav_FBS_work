```python
def findoutliers(df,col):
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.boxplot(df[col])
    plt.subplot(1,2,2)
    mean1=df[col].mean()
    std=df[col].std()
    sns.histplot(data=df,x=col,kde=True)
    plt.axvline(mean1,color="red")
    plt.axvline(mean1+3*std)
    plt.axvline(mean1-3*std)
    return df[df[df[col] < mean1+3*std] | df[df[col]>mean1-3*std]]
```

```python
findoutliers(df,"release_year")
```
