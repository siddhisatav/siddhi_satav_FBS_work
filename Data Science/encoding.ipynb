```python
data = pd.read_csv("cars93.csv")
data.head()
```

```python
data.isna().sum()
```

```python
data["AirBags"]=data["AirBags"].fillna(data["AirBags"].mode()[0])
```

```python
data["Luggage.room"]=data["Luggage.room"].fillna(data["Luggage.room"].mean())
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data["Type"] = le.fit_transform(data["Type"])
```

```python
data.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
od=OrdinalEncoder(categories=[['Compact', 'Small', 'Midsize','Sporty', 'Large', 'Van']])
```

```python
data["Type"]=od.fit_transform(data[["Type"]]).astype('int')
```

```python
data.head()
```

```python
oh = OneHotEncoder()
result = oh.fit_transform(data[["AirBags"]]).toarray()
result=pd.DataFrame(result , columns=["Driver only","Driver & passenger","none"])
data = pd.concat([data,result],axis = 1)
data.drop("AirBags",axis = 1, inplace=True )
```

```python
data.head()
```

```python
data = pd.get_dummies(data,columns=["AirBags"],dtype="int")
```
