```python
age__group=pd.DataFrame(columns=["20-30","31-40","41-50","50+"])
age__group
```

```python
def fun1(age):
    if(age > 20 and age <30):
        return "20-30"

    elif(age>31 and age<40):
        return "31-40"

    elif(age>41 and age<50):
        return "41-50"
    else:
        return "50+"

    
```

```python
import numpy as np

dt["age_group"] = dt["Age"].apply(fun1)
dt.head()
```

```python
dt["age_group"].value_counts()
```

```python
dt[dt["age_group"] =="50+"]
```
