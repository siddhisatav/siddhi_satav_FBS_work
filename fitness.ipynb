```python
q31 = df["sleep_duration(hours)"].quantile(0.75)
q11 = df["sleep_duration(hours)"].quantile(0.25)
iqr1=q31-q11
up_sleep= q31 + (iqr1*1.5)
print(up_sleep)
lp = q11-(iqr1 * 1.5)
print(lp)
```

```python
df[(df["sleep_duration(hours)"] > up_sleep) | (df["sleep_duration(hours)"] < lp )]
```

```python
 df.loc[df["sleep_duration(hours)"] > up_sleep, "sleep_duration(hours)"] = 8.25
```

```python
 df.loc[df["sleep_duration(hours)"] < lp, "sleep_duration(hours)"] = 6.25
```
