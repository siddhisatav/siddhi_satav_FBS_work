```python
import pandas as pd

data = {
    "Order ID": [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008],
    "Customer": ["Alice", "Bob", "Charlie", "Alice", "David", "Charlie", "Bob", "Alice"],
    "Product": ["Laptop", "Phone", "Chair", "Table", "Laptop", "Phone", "Chair", None],
    "Category": ["Electronics", "Electronics", "Furniture", "Furniture", "Electronics", "Electronics", "Furniture", None],
    "Quantity": [2, 1, 4, 1, 1, 2, 2, None],
    "Price": [800, 500, 100, 250, 800, 500, 100, None],
    "Total Sales": [1600, 500, 400, 250, 800, 1000, 200, None]
}

df = pd.DataFrame(data)

print(df)
```

```python
df["Discount"]=df["Total Sales"] * 0.10
```
